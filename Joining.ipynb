```python
def transcript_binary_search(data, val, end=False):
    lo, hi = 0, len(data) - 1
    best_ind = lo
    if end:
        while lo <= hi:
            mid = lo + (hi - lo) // 2
            if data[mid]['start'] + data[mid]['duration'] < val:
                lo = mid + 1
            elif data[mid]['start'] + data[mid]['duration'] > val:
                hi = mid - 1
            else:
                best_ind = mid
                break
            # check if data[mid] is closer to val than data[best_ind] 
            if abs(data[mid]['start'] + data[mid]['duration'] - val) < abs(data[best_ind]['start'] + data[best_ind]['duration'] - val):
                best_ind = mid
    else:
        while lo <= hi:
            mid = lo + (hi - lo) // 2
            if data[mid]['start'] < val:
                lo = mid + 1
            elif data[mid]['start'] > val:
                hi = mid - 1
            else:
                best_ind = mid
                break
            # check if data[mid] is closer to val than data[best_ind] 
            if abs(data[mid]['start'] - val) < abs(data[best_ind]['start'] - val):
                best_ind = mid
    return best_ind

def transcript_search(transcript, start, end):
    starting_index = transcript_binary_search(transcript, start)
    ending_index = transcript_binary_search(transcript, end, end=True)
    text_list = [x['text'] for x in transcript[starting_index:ending_index+1]]
    return text_list

def apply_func(row):
    video = row['videoID']
    start = row['startTime']
    end = row['endTime']
    
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video)
    except Exception as e:
        return None, None
    
    text_list = transcript_search(transcript, start, end)
    return ' '.join(text_list), ' '.join([sentence['text'] for sentence in transcript])
```

```python
structured_data = []

for index, transcript in enumerate(transcripts):
    if transcript[1] is None:
        continue
    
    video = transcript[0]
    for sentence in transcript[1]:
        new_row = {'videoID': video, 'start': sentence['start'], 'duration': sentence['duration'], 'text': sentence['text']}
        structured_data.append(new_row)
        
    print(index, end='\r')
    
transcript_df = pd.concat([transcript_df, pd.DataFrame(structured_data)])
```

```python
merged_df = pd.merge(df, transcript_df, how='left', on='videoID')
```

```python
merged_df = merged_df.drop(['votes', 'locked', 'incorrectVotes', 'userID', 'timeSubmitted', 'views', 'videoDuration', 'hidden', 'reputation', 'shadowHidden', 'hashedVideoID', 'userAgent', 'description'], axis=1)
merged_df
```

```python
def is_sponsored(row):
    if row['startTime'] <= row['start'] <= row['endTime']:
        return True
    elif row['startTime'] <= row['start'] + row['duration'] <= row['endTime']:
        return True
    
    return False
```

```python
import swifter

merged_df['sponsored'] = merged_df.swifter.apply(is_sponsored, axis=1)
```

```python
merged_df.head(10)
```

```python
merged_df = merged_df.sort_values('sponsored', ascending=False).drop_duplicates(subset=['videoID', 'start', 'duration', 'text']).sort_values(['videoID', 'start', 'duration']).reset_index(drop=True)
```

Text cleaning time!

Removing stopwords is can provide different results for topic modeling, so I'll make a column without the stopwords

```python
merged_df['text'] = merged_df['text'].replace('', np.nan)
merged_df = merged_df[~(merged_df['text'].isna())]
```

```python
import re


def clean_text(text):
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove symbols (keep letters, numbers, spaces)
    text = unicodedata.normalize('NFKC', text)  # Remove weird space characters
    return text.lower()

merged_df['text'] = merged_df['text'].swifter.apply(clean_text)
merged_df['text'].replace(r'^\s*$', pd.NA, regex=True, inplace=True)
merged_df = merged_df[~(merged_df['text'].isna())]
```

```python
from nltk.corpus import stopwords

stop = stopwords.words('english')

# Remove stop words from text and make the result a new column
merged_df['text_without_stopwords'] = merged_df['text'].swifter.apply(lambda x: ' '.join([word for word in x.split() if word not in stop]))
```

```python
merged_df = merged_df.drop(['startTime', 'endTime', 'actionType', 'service'], axis=1)

# Remove UUID from unsponsored sentences
merged_df.loc[merged_df['sponsored'] == False, 'UUID'] = np.nan


merged_df
```

Selecting the train and test sets

```python
from sklearn.model_selection import train_test_split

video_ids = merged_df['videoID'].unique()

training_ids, testing_ids = train_test_split(video_ids, test_size=0.2, random_state=36)

train_df = merged_df[merged_df['videoID'].isin(training_ids)]
test_df = merged_df[merged_df['videoID'].isin(testing_ids)]
```
